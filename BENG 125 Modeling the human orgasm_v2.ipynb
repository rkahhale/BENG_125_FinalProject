```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = [0]
v = [0]
t = np.arange(0,5,.000001)
squigly = -.000001
alpha = 1
b = 1
c = 1
g = 1

for i in range(0,14):
    u.append(u[i]+ ((2*squigly*u[i]) + (alpha*v[i])))
    omega = np.sin(t[i])
    v.append(v[i]+ ((b*v[i]) + (c*(u[i]**3)) + (g*np.sin(omega*t[i]))))
print(u)
print(v)
plt.plot(u,v)
plt.show()
#goes from being logarithmic to liniear at 13
```

```python
u = [0]
v = [0]
t = np.arange(0,500)
m1 = .1
m2 = .1
m3 = .1
m4 = .1
for i in range(0,500): 
    u.append(u[i]+ ((-1*m1*u[i])-(m2*v[i])+np.sin(t[i])))
    v.append(v[i]+ ((-1*m3*u[i])-(m4*v[i])))
plt.plot(u,v)
plt.show()
#goes from being logarithmic to liniear at 13
```
